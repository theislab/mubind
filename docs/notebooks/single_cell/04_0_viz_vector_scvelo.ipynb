```python

# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Creating dataset
n_points = 3
x = np.arange(0, n_points)
y = np.arange(0, n_points)

# Creating grids
X, Y = np.meshgrid(x, y)

# x-component to the right
u = np.ones((n_points, n_points))

# y-component zero
v = np.zeros((n_points, n_points))

fig = plt.figure(figsize = (12, 7))

# Plotting stream plot
plt.streamplot(X, Y, u, v, density = 0.5)

# show plot
print(X)
print(Y)
plt.show()
```
