```python
base_for = "ACGT"
base_rev = "TGCA"
comp_tab = str.maketrans(base_for, base_rev)
```

```python
pd.DataFrame(model.conv_mono[2].weight.squeeze().cpu().detach().numpy())
```

```python
fwd = 'TTGTATTGACGT'
rev = fwd.translate(comp_tab)[::-1]
# variants up to one mismatch
n_mut = 1
kmers = [s[:i] + nt + s[i + 1:] for s in [fwd, rev] for i in range(len(s))  for nt in 'ATCG']
print(len(kmers))
for curr_mut in range(n_mut):
    next_kmers = []
    for k in kmers:
        next_kmers += [s[:i] + nt + s[i + 1:] for s in [k] for i in range(len(s))  for nt in 'ATCG']
    kmers = next_kmers
print(len(kmers))
```

```python
data['found'] = False
for i, s in enumerate(kmers):
    if i % 100 == 0:
        print(i, s)
    data['found'] = np.where(data.seq.str.contains(s), True, data['found'])
```

```python
data[data['found']].to_csv('../data/countTable.0.CTCF_r3_2nd_binding_mode_subset.tsv.gz')
```

### Calculation of kmer enrichments using ProBoundTools

```python
df = pd.DataFrame({'seq': pred[0], 't0': pred[1][:, 0], 't1': pred[1][:, 1], 'p0': pred[2][:, 0], 'p1': pred[2][:, 1]})
df.sort_values('seq')
```

```python
model_pred = pd.read_csv('../data/ctcf_r3_count_pred.txt.gz', header=None, sep='\t')
model_pred.columns = ['seq', 0, 1]
model_pred
```

```python
y_pred, y_true = model_pred[model_pred.columns[1:]].to_numpy(), model_pred[model_pred.columns[1:]].to_numpy()

# this is the loss as calculated using the model
np.mean(y_pred.flatten() - y_true.flatten() * np.log(y_pred.flatten()))
```

```python
# I copied most of the code from mb.pl.kmer_enrichment, because the function expects 
# a model and a dataset as parameters and I didn't want to change that.
```

```python
pred_kmers = mb.tl.seqs2kmers(model_pred['seq'], k=8, counts=np.array(model_pred[[0, 1]]))
pred_labels = ['p'+str(i) for i in range(train.dataset.n_rounds + 1)]
pred_kmers[pred_labels] = np.stack(pred_kmers['counts'].to_numpy())
```

```python
target_kmers = mb.tl.seqs2kmers(df['seq'], k=8, counts=np.array(df[['t0', 't1']]))
target_labels = ['t'+str(i) for i in range(train.dataset.n_rounds + 1)]
target_kmers[target_labels] = np.stack(target_kmers['counts'].to_numpy())
```

```python
base_round = 0
enr_round = -1
counts = target_kmers[target_labels].merge(pred_kmers[pred_labels], left_index=True, right_index=True, how='outer').fillna(0)
if enr_round == -1:
    enr_round = train.dataset.n_rounds
counts['enr_pred'] = (1 + counts[pred_labels[enr_round]]) / (1 + counts[pred_labels[base_round]])
counts['enr_obs'] = (1 + counts[target_labels[enr_round]]) / (1 + counts[target_labels[base_round]])
counts['f_pred'] = (1 / (enr_round - base_round)) * np.log10(counts['enr_pred'])
counts['f_obs'] = (1 / (enr_round - base_round)) * np.log10(counts['enr_obs'])

# plotting
p = sns.displot(counts, x='enr_pred', y='enr_obs', cbar=True)
p.set(xscale='log', yscale='log')
plt.plot([0.1, 10], [0.1, 10], linewidth=2)

print('R^2:', r2_score(counts['f_obs'], counts['f_pred']))
plt.show()

# grafic shows prediction of the Probound model via ProboundTools
```
