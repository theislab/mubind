```python
n_rounds = 1

df = mb.bindome.datasets.ProBound.ctcf()
df = df.sort_values(1, ascending=False).reset_index(drop=True)

data = pd.read_csv('../data/countTable.0.CTCF_r3.tsv.gz', sep='\t', header=None)
data.columns = ['seq'] + [i for i in range(n_rounds+1)]
# data = data.sample(n=1000)
# data.index = range(len(data))
dataset = mb.datasets.SelexDataset(data, n_rounds=n_rounds)
train = tdata.DataLoader(dataset=dataset, batch_size=256, shuffle=True)

# data = df.head(1000)
data = df.sample(n=100)
# data = df.copy()

set(data[0])

n_rounds = 1

from matplotlib import rcParams
rcParams['figure.figsize'] = 5, 1

print('loading object (# entries)', data.shape[0])
dataset = mb.datasets.SelexDataset(data, n_rounds=n_rounds)
train = tdata.DataLoader(dataset=dataset, batch_size=256, shuffle=True)
```

```python
choice_w = [15, 16]
choice_lr = [0.01, 0.001]
choice_n_kernels = [3]
choice_n_epochs = [100]
choice_weight_decay = [0.001]
choice_batch_size = [128, 256, 512]
```

```python

seed=2
def objective(params):
    w = int(params['w'])
    lr = params['lr']
    num_epochs = params['num_epochs']
    n_kernels = params['n_kernels']
    weight_decay = params['weight_decay']
    batch_size = params['batch_size']
    dirichlet_regularization = 0
        
    print('here...', [n_kernels, w, lr, num_epochs, batch_size])
    # msl = int(params['min_samples_leaf'])
    # mss = int(params['min_samples_split'])
    # print('loading object (# entries)', data.shape[0])
    dataset = mb.datasets.SelexDataset(data, n_rounds=n_rounds)
    train = tdata.DataLoader(dataset=dataset, batch_size=batch_size, shuffle=True)

    model, loss = mb.tl.train_iterative(train, device, w=w, show_logo=0, optimize_motif_shift=True,# criterion=mb.tl.ProboundLoss(),
                                    dirichlet_regularization=dirichlet_regularization, # 10 ** dirichlet_regularization_log,
                                    lr=[lr] * n_kernels, weight_decay=[weight_decay] * n_kernels, ignore_kernel=1, verbose=0,
                                    num_epochs=num_epochs, early_stopping=50, use_dinuc=False, # optimiser=torch.optim.LBFGS,
                                    n_kernels=n_kernels, log_each=50, stop_at_kernel=None) #  seed=seed) # seeds.index[0]) # 
    return loss

def optimize(trial):
    params={'w': hp.choice('w', choice_w),
            'lr': hp.choice('lr', choice_lr),
            'n_kernels': hp.choice('n_kernels', choice_n_kernels),
           'num_epochs': hp.choice('num_epochs', choice_num_epochs),
           'weight_decay': hp.choice('weight_decay', choice_weight_decay),
            'batch_size': hp.choice('batch_size', choice_batch_size)}
    best=fmin(fn=objective, space=params, algo=tpe.suggest, trials=trial,
              max_evals=10, rstate=np.random.default_rng(seed))
    return best

trial = Trials()
best = optimize(trial)
```
